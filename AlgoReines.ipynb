```python
def placer_reine(ListReines):
    #fonction qui nous donne les cases libres
    #(répértorier les cases qui sont accéssibles dans la ligne
    #& va garder uniquement les cases valides

    lstPossible =[(len(ListReines),i) for i in range(8)]
    #générer directement toutes les cases d'une ligne;
    #[(0,0),(0,1),...,(N°reine,i),(ligne,colonne)]
    #avec ; i de 0 à 7
    ListValide = [] #initialisation du liste des cases valides
    
    for i in lstPossible: # i : index d'une case d'un ligne 
        testValide = True #initialisation du test de validation
        for j in ListReines: # j : index d'une reine dans la liste
            if i[1] == j[1]: #validité de colonne ; i=[0,1,2],j=[0,1,2]
                testValide = False
            elif abs((i[0]-j[0])/(i[1]-j[1]))==1: #validité de diagonale (pente)
                testValide = False
        if testValide:
            ListValide.append(i) #ajouter la case 'i' au liste des cases valides
    return ListValide #afficher la liste des cases valides
```

```python
def placer_toutes_reines(ListReines=[]):
    #ListReines: liste des reines déja placées
    #( ([]) initialisation ; vide )
    #cette fonction va parcourir la liste 
    #des emplacements libres de prochaines lignes de l'échiquier
    for i in placer_reine(ListReines):
        if len(ListReines)== 7:
            print(ListReines+[i])
        else:
            placer_toutes_reines(ListReines+[i]) 

placer_toutes_reines()
```
